```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fractions import Fraction
```

```python
data_8 = pd.read_csv('Rand_ball_4_8.txt', sep=" ", header=None)
```

```python
data_8.columns = ["vert", "vol"]
```

```python
hist_8 = data_8.hist(column='vol', bins = 12, range=[3.5, 4.1], figsize=(12,8), grid=False, color='#86bf91', zorder=2, rwidth=0.9);
hist_8 = hist_8[0]
for x in hist_8:
    # Remove title
    x.set_title("")

    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    x.grid(axis='y')

    # Set x-axis label
    x.set_xlabel("Volume", labelpad=10, weight='bold', size=16)

    # Set y-axis label
    x.set_ylabel("Frequency", labelpad=10, weight='bold', size=16)
    
    x.set_ylim([0, 100])
    
    for tick in x.xaxis.get_major_ticks():
        tick.label.set_fontsize(16)
    for tick in x.yaxis.get_major_ticks():
        tick.label.set_fontsize(16)
    
```

```python
fig_8 = hist_8[0].get_figure()
fig_8.savefig('hist_4_8.eps', format='eps', bbox_inches='tight')
```
